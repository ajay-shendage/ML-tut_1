```python
# This is highly imbalanced

# 0 -> legit
# 1 -> fraud
```

```python
# separating the legit and fraud transcatioms

legit = df[df.Class==0]
fraud = df[df.Class==1]
```

```python
# Under Sampling

# build a sample dataset containg similar distribution of legit and fraud transactions
```

```python
legit_sample = legit.sample(n=492)
```

```python
# Contatenate two DataFrames

new_df = pd.concat([legit_sample,fraud],axis=0)
```
