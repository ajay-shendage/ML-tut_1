```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Label Encoding of Breast Cancer Dataset

df = pd.read_csv(r"C:\Users\ajays\OneDrive\Desktop\ML\datasets\breast_cancer_data.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=['Unnamed: 32'])
```

```python
df['diagnosis'].value_counts() 
```

```python
# Load the label encoder function

label = LabelEncoder()

labels = label.fit_transform(df.diagnosis)
```

```python
# Appending the labels into dataframe

df['target']=labels
```
