```python
# import libraries

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# seaborn has in built datasets

# total bill vs tip dataset

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# setting a theme for plot

sns.set_theme()
```

```python
# visualize the data

sns.relplot(data=tips, x='total_bill', y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# house price dataset

from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
house = pd.DataFrame(housing.data,columns=housing.feature_names)
house['price'] = housing.target
house.head()
```

```python
# Distribution plot

sns.histplot(house['price'],kde=True,kde_kws=dict(cut=3),alpha=.4, edgecolor=(1, 1, 1, .4))
```

```python
sns.kdeplot(house['price'])
```

```python
# Correlation 

cor = house.corr()
```

```python
# Heat Map

# construction a heat map

plt.figure(figsize=(10,10))
sns.heatmap(cor,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
